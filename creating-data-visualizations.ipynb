```python
import csv

rows = []
with open("Datasets/Attack_Dataset.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        # print(row)
        rows.append(row)

# print(rows)

rows = rows[1:]
dic = {}
# print(rows[0])

for row in rows:
    # print(row[3].strip())
    if row[3].strip() not in dic:
        dic[row[3].strip()] = 0
    dic[row[3].strip()] += 1

print(dic)



```
