```python
import csv
import matplotlib.pyplot as plt

rows = []

with open("Datasets/Attack_Dataset.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        # print(row)
        rows.append(row)

# print(rows)

rows = rows[1:]
dic = {}
# print(rows[0])

for row in rows:
    # print(row[3].strip())
    if row[3].strip() not in dic:
        dic[row[3].strip()] = 0
    dic[row[3].strip()] += 1


# for key, value in dic.items():
#     print(key, ":", value)
dic['other'] = 0
keys = []
for key, value in dic.items():
    if value <= 20 :
        dic['other'] += 1
        keys.append(key)

for key in keys:
    del dic[key]
del dic['other']


print(dic)
plt.figure(figsize=(20, 6)) # width=12, height=6
plt.bar(dic.keys(),dic.values())
plt.xlabel("Attack Type")
plt.ylabel("Count")
plt.title("Most Common Attack Types")
plt.xticks(rotation=45, ha="right", fontsize=11)
plt.tight_layout()
plt.show()




```
