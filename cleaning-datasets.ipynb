```python
import pandas as pd
import numpy as np



_1_attacks = pd.read_csv('Datasets/Attack_Dataset.csv')
_2_india = pd.read_csv('Datasets/cybersecurity_cases_india_combined.csv')
_3_globald = pd.read_csv('Datasets/Global_Cybersecurity_Threats_2015-2024.csv')
_4_vulnerability = pd.read_csv('Datasets/vulnerability_assessment_data.csv')

# subset so specify only drop if these specific columns are missing
_1_attacks = _1_attacks.dropna(subset=['ID','Title','Scenario Description','Tools Used','Attack Steps ','Impact','Detection Method','Solution','Tags','Source'])
_2_india = _2_india.dropna(subset=['Day','Amount_Lost_INR','City']) # drops rows with missing data
_3_globald = _3_globald.dropna(subset=['Country', 'Financial Loss (in Million $)', 'Number of Affected Users' , 'Attack Source' , 'Defense Mechanism Used' , 'Incident Resolution Time (in Hours)'])
_4_vulnerability = _4_vulnerability.dropna(subset=['ID', 'Description', 'Affected Software' , 'Discovered Date' , 'Patch Date' , 'Description Length', 'Days to Patch'])

# _1_attacks.head()
# _2_india.head()
# _3_globald.head()
# _4_vulnerability.head()
```
